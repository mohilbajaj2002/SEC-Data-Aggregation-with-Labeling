```python
file = open('Event Classification/all_events_2018.txt', 'r').read()
text = file.split('\n\n')

score_mna = score_npl = score_negatives = times = score_final = count = 0
text_score = []
event_type = []

m_a = {'merger': 1, 'acquisition': 1, 'deal': 1, 'transaction': 1, 'sale': 1, 'acquire': 1}
npl = {'released': 1, 'release': 1, 'launch': 1, 'launched': 1, 'product': 1, 'products': 1, 'offering': 1}
nme = {}
negatives = {'officers': -1, 'officer': -1, 'president': -1, 'recruitment': -1, 'role': -1, 'serves': -1, 'served': -1, 
             'responsible': -1, 'appointed': -1, 'employed': -1, 'stock': -1, 
             'revenue': -1, 'revenues': -1, 'accounting': -1, 'standard': -1,'standards': -1,
             'consolidated': -1, 'financial': -1, 'statements': -1, 
             'assets': -1, 'liabilities': -1, 'tax': -1, 'taxes': -1, 'deferred': -1,'fiscal': -1, 
             'cash': -1, 'unearned': -1, 'receivable': -1, 'infringement': -1, 'settlement': -1,
             'receivables': -1, 'accounts': -1, 'lease': -1, 'leases': -1, 'balance': -1, 'sheet': -1,
             'tangible': -1, 'intangible': -1, 'impairment': -1, 
             'collateral': -1, 'securities': -1, 'currency': -1, 'options': -1,'hedging': -1,
             'hedged': -1, 'contracts': -1, 'instruments': -1,'futures': -1,
              'put': -1,'call': -1, 'swap': -1, 'swaps': -1, 'credit': -1, 'default': -1, 'goodwill': -1, 
             'amortisation': -1, 'unsecured': -1, 'debt': -1,
             'accrued': -1, 'defendant': -1, 'plaintiffs': -1, 'evidence': -1, 'appeal': -1, 'appealed': -1, 
             'court': -1, 'filed': -1, 'complaint': -1, 'litigation': -1}

for i in range (0, len(text)):
    for m in m_a:
        if m in text[i]:
            times = text[i].lower().split().count(m)
            score_mna = score_mna + (m_a[m]*times)
            
    times = 0
    for n in npl:
        if n in text[i]:
            times = text[i].lower().split().count(n)
            score_npl = score_npl + (npl[n]*times)
            
    times = 0
    for n in negatives:
        if n in text[i]:
            times = text[i].lower().split().count(n)
            score_negatives = score_negatives + (negatives[n]*times)
    
    score_final = score_mna + score_npl + score_negatives
    text_score.append(score_final)
    
    if (score_final >= 0 and score_mna > score_npl):
        classification = 'M & A'
        event_type.append(0)
    elif (score_final >= 0 and score_npl > score_mna):
        classification = 'New Product Launch'
        event_type.append(1)
    elif (score_final >= 0 and score_npl == score_mna):
        classification = 'More Analysis Needed'
        event_type.append(3)
    else:
        classification = 'Other'
        event_type.append(2)
        
    count = count + 1
    print (text[i])
    print ('Event No: ', count, ', Score: ', score_final, ', Classification: ', classification)
    print ('')
    score_mna = score_npl = score_negatives = times = score_final = 0
```
